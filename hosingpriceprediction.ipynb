```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
# simple data set with only two data points-
# a house with 1000 sqft sold for 300000 dollars (1sqft-300(1000s of dollars))
# a house with 2000 sqft sold for 500000 dollars (2sqft-500)
# want to fit a linear regression model through these two points
# predict price for other houses
```

```python
x_feature = np.array([1.0, 2.0])
y_target  = np.array([300, 500]) 
```

```python
# Let m is the number of training examples
m = len(x_feature)
print(f"Number of training examples is: {m}")
```

```python
# (x^{(i)},y^{(i)}) is the i^{th} training example. 
i = 0
x_i = x_feature[i]
y_i = y_target[i]
print(f"(x^({i}),y^({i})) = ({x_i}, {y_i})")
```

```python
i = 1
x_i = x_feature[i]
y_i = y_target[i]
print(f"(x^({i}),y^({i})) = ({x_i}, {y_i})")
```

```python
# Using thr scatter() function in matplotlib
# The function arguments marker and c show the points as red crosses 
# plot the data points
plt.scatter(x_feature, y_target, marker='x', c='r')
# Setting the y-axis
plt.ylabel('Price (in 1000s of dollars)')
#setting the x-axis
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
# f_wb = w * x[i] + b
def calc_model_output(w,b,x):
    m = x.shape
    f_wb = np.zeros(m)
    for i in range(len(x)):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb = calc_model_output(w, b, x_feature)
#plotting model prediction
plt.plot(x_feature, tmp_f_wb, c='b', label='Our Prediction')
#plotting data points
plt.scatter(x_feature, y_target, marker='x', c='r', label='Actual Values')
#set the title
plt.title("Housing Prices")
#set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
#setting the x-axis
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# conclusion - setting w = 100 and b = 100 does not result in a line that fits our data
```

```python
# For w=200 and b=100 we get:
w = 200
b = 100
def calc_model_output(w,b,x):
    m = x.shape
    f_wb = np.zeros(m)
    for i in range(len(x)):
        f_wb[i] = w * x[i] + b
    return f_wb
tmp_f_wb = calc_model_output(w, b, x_feature)
#plotting model prediction
plt.plot(x_feature, tmp_f_wb, c='b', label='Our Prediction')
#plotting data points
plt.scatter(x_feature, y_target, marker='x', c='r', label='Actual Values')
#set the title
plt.title("Housing Prices")
#set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
#setting the x-axis
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# for w= 200 and b = 100 results in a line that fits our data
```

```python
# Prediction - predict the price of a house with 1200 sqft
w = 200
b = 100
cost_1200sqft = w * 1.2 + b
print(f"${cost_1200sqft:.0f} thousand dollars")
```
